## Test Case 1: AI Healthcare Grant Proposal

```python
# Test data
grant_name_1 = "NSF AI for Healthcare Innovation"
match_criteria_1 = "AI/ML applications in healthcare, interdisciplinary collaboration"
eligibility_1 = "Assistant or Associate Professor, STEM field"
match_justification_1 = "Strong alignment with AI/ML expertise and healthcare applications. Candidate has relevant publications and interdisciplinary experience."

sourced_data_1 = {
    "experience": ["Assistant Professor in Computer Science", "5 years industry experience in healthcare tech"],
    "publications": ["Machine Learning in Healthcare", "AI Ethics Framework"],
    "expertise": ["Machine Learning", "Artificial Intelligence", "Healthcare Informatics"],
    "credentials": ["PhD Computer Science", "NSF CAREER Award"]
}

collaborators_1 = [
    {"name": "Dr. Sarah Chen", "department": "Computer Science", "expertise": "Machine Learning, NLP"},
    {"name": "Dr. Michael Rodriguez", "department": "Biomedical Engineering", "expertise": "Medical Imaging, AI"},
    {"name": "Dr. Lisa Park", "department": "Public Health", "expertise": "Epidemiology, Health Informatics"}
]

print("Test Case 1: AI Healthcare Grant")
print(f"Grant: {grant_name_1}")
print(f"Match Criteria: {match_criteria_1}")
print(f"Collaborators: {len(collaborators_1)}")
print("\n" + "="*50 + "\n")
```

```python
async def test_drafting_agent(grant, criteria, elig, justification, sourced, collabs):
    results = []
    async for message in run_drafting_agent(grant, criteria, elig, justification, sourced, collabs):
        results.append(message)
        print(f"[{message['agent']}] {message['step']}")
        if message['done']:
            print("\n" + "="*70)
            print("PROPOSAL DRAFT")
            print("="*70 + "\n")
            print(message['output']['proposalDraft'])
            print("\n" + "="*70)
            
            # Word count
            word_count = len(message['output']['proposalDraft'].split())
            print(f"\nWord Count: {word_count}")
    return results

results_1 = await test_drafting_agent(
    grant_name_1, match_criteria_1, eligibility_1, 
    match_justification_1, sourced_data_1, collaborators_1
)
```

## Analyze Draft Quality

```python
# Analyze proposal characteristics
print("Draft Analysis:\n")

for i, results in enumerate([results_1, results_2, results_3], 1):
    final_output = [r for r in results if r['done']][0]['output']
    draft = final_output['proposalDraft']
    
    word_count = len(draft.split())
    sections = draft.count('##')
    has_intro = 'Introduction' in draft
    has_objectives = 'Objectives' in draft
    has_methodology = 'Methodology' in draft
    has_impact = 'Impact' in draft
    
    print(f"Test Case {i}:")
    print(f"  Word Count: {word_count}")
    print(f"  Sections: {sections}")
    print(f"  Has Introduction: {has_intro}")
    print(f"  Has Objectives: {has_objectives}")
    print(f"  Has Methodology: {has_methodology}")
    print(f"  Has Impact: {has_impact}")
    print(f"  Target Range (300-500): {'✓' if 300 <= word_count <= 500 else '✗'}")
    print()
```
