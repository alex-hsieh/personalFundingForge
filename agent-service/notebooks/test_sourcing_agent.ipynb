# SourcingAgent Test Notebook

This notebook tests the SourcingAgent which extracts and structures user CV and profile data.

**Model**: Claude Haiku (anthropic.claude-haiku-4-5-20251001-v1:0)

**Purpose**: Source and structure user experience, publications, and expertise

```python
import sys
import os
import asyncio
import json

# Add parent directory to path
sys.path.append(os.path.dirname(os.path.abspath('')))

from agents.sourcing import run_sourcing_agent
```

## Test Case 1: Assistant Professor Profile

```python
# Sample user profile
user_profile_1 = {
    "role": "Assistant Professor",
    "year": "2023",
    "program": "Computer Science"
}

print("Testing SourcingAgent with profile:")
print(json.dumps(user_profile_1, indent=2))
print("\n" + "="*50 + "\n")
```

```python
# Run the agent and collect results
async def test_sourcing_agent(profile):
    results = []
    async for message in run_sourcing_agent(profile):
        results.append(message)
        print(f"[{message['agent']}] {message['step']}")
        if message['done']:
            print("\nFinal Output:")
            print(json.dumps(message['output'], indent=2))
    return results

results_1 = await test_sourcing_agent(user_profile_1)
```

## Test Case 2: Postdoctoral Researcher Profile

```python
user_profile_2 = {
    "role": "Postdoctoral Researcher",
    "year": "2024",
    "program": "Biomedical Engineering"
}

print("Testing SourcingAgent with profile:")
print(json.dumps(user_profile_2, indent=2))
print("\n" + "="*50 + "\n")

results_2 = await test_sourcing_agent(user_profile_2)
```

## Test Case 3: Associate Professor Profile

```python
user_profile_3 = {
    "role": "Associate Professor",
    "year": "2020",
    "program": "Environmental Science"
}

print("Testing SourcingAgent with profile:")
print(json.dumps(user_profile_3, indent=2))
print("\n" + "="*50 + "\n")

results_3 = await test_sourcing_agent(user_profile_3)
```

## Analyze Results

```python
# Compare outputs
print("Summary of all test cases:\n")

for i, results in enumerate([results_1, results_2, results_3], 1):
    final_output = [r for r in results if r['done']][0]['output']
    print(f"Test Case {i}:")
    print(f"  Experience items: {len(final_output.get('experience', []))}")
    print(f"  Publications: {len(final_output.get('publications', []))}")
    print(f"  Expertise areas: {len(final_output.get('expertise', []))}")
    print(f"  Credentials: {len(final_output.get('credentials', []))}")
    print()
```
