## Test Case 1: AI/ML Healthcare Grant

```python
user_profile_1 = {
    "role": "Assistant Professor",
    "year": "2023",
    "program": "Computer Science"
}

match_criteria_1 = "AI/ML applications in healthcare, interdisciplinary collaboration"

print("Test Case 1: AI/ML Healthcare Grant")
print("User Profile:", json.dumps(user_profile_1, indent=2))
print(f"Match Criteria: {match_criteria_1}")
print("\n" + "="*50 + "\n")
```

```python
async def test_collaborator_agent(faculty, criteria, profile):
    results = []
    async for message in run_collaborator_agent(faculty, criteria, profile):
        results.append(message)
        print(f"[{message['agent']}] {message['step']}")
        if message['done']:
            print("\nRecommended Collaborators:")
            collaborators = message['output']['collaborators']
            for i, collab in enumerate(collaborators, 1):
                print(f"\n{i}. {collab['name']}")
                print(f"   Department: {collab['department']}")
                print(f"   Expertise: {collab['expertise']}")
                print(f"   Relevance Score: {collab['relevanceScore']}%")
    return results

results_1 = await test_collaborator_agent(faculty_list, match_criteria_1, user_profile_1)
```

## Test Case 3: Interdisciplinary HCI Grant

```python
user_profile_3 = {
    "role": "Assistant Professor",
    "year": "2024",
    "program": "Psychology"
}

match_criteria_3 = "Human-computer interaction, cognitive systems, user experience"

print("Test Case 3: Interdisciplinary HCI Grant")
print("User Profile:", json.dumps(user_profile_3, indent=2))
print(f"Match Criteria: {match_criteria_3}")
print("\n" + "="*50 + "\n")

results_3 = await test_collaborator_agent(faculty_list, match_criteria_3, user_profile_3)
```

## Analyze Collaboration Patterns

```python
# Analyze department diversity and relevance scores
print("Collaboration Analysis:\n")

for i, results in enumerate([results_1, results_2, results_3], 1):
    final_output = [r for r in results if r['done']][0]['output']
    collaborators = final_output['collaborators']
    
    departments = [c['department'] for c in collaborators]
    avg_score = sum(c['relevanceScore'] for c in collaborators) / len(collaborators)
    
    print(f"Test Case {i}:")
    print(f"  Departments: {', '.join(departments)}")
    print(f"  Average Relevance Score: {avg_score:.1f}%")
    print(f"  Interdisciplinary: {'Yes' if len(set(departments)) > 1 else 'No'}")
    print()
```
